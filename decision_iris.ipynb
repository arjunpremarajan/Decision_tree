```python
import pandas as pd 
import numpy as np
df=pd.read_csv('/home/arjun/Downloads/Iris.csv')
df.head()
```

```python
df=df.drop(['Id'],axis=1)
x=df.iloc[0:,:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
x_train=scalar.fit_transform(x_train)
x_test=scalar.transform(x_test)
x_train
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier() #default=gini index
dt.fit(x_train,y_train)
y_pred=dt.predict(x_test)
y_pred
```

```python
#plot tree
from sklearn import tree
import matplotlib.pyplot as plt
plt.figure(figsize=(15,15))
tree.plot_tree(dt,feature_names=['sepal_l','sepal_w','petal_l','petal_w'],class_names=['iris_versi','iris_virgin','iris_set'],filled=True)
```

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
acc=accuracy_score(y_test,y_pred)
print(acc)
```

```python
confu=confusion_matrix(y_test,y_pred)
print(confu)
```

```python
classi=classification_report(y_test,y_pred)
print(classi)
```
